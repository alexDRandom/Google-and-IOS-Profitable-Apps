```python
freeapps = dfgoogle[dfgoogle['Type'] == 'Free']

print(freeapps)
```

```python
freeapps.head(10)
```

```python
duplicados = freeapps[freeapps['App'].duplicated(keep=False)]  # Mantiene todos los duplicados

# Contar los duplicados
num_duplicados = freeapps['App'].duplicated().sum()

# Mostrar los duplicados
print("Número de aplicaciones duplicadas:", num_duplicados)
print("Aplicaciones duplicadas:\n", duplicados[['App']])
```

```python
# Contar cuántas veces aparece cada aplicación
conteo_apps = freeapps['App'].value_counts()

# Filtrar solo las que tienen más de una aparición
duplicados_apps = conteo_apps[conteo_apps > 1]

print("Aplicaciones duplicadas y sus conteos:\n", duplicados_apps)
```

```python
# Verificar el tipo de la columna 'Reviews'
print("Tipo de la columna 'Reviews':", freeapps['Reviews'].dtype)

# Identificar valores no numéricos
no_numericos = freeapps[~freeapps['Reviews'].str.isnumeric()]

# Mostrar los valores no numéricos si existen
if not no_numericos.empty:
    print("Valores no numéricos en la columna 'Reviews':\n", no_numericos[['App', 'Reviews']])
else:
    print("Todos los valores en la columna 'Reviews' son numéricos.")
```

```python
# Asegúrate de que la columna 'Reviews' sea numérica
freeapps['Reviews'] = pd.to_numeric(freeapps['Reviews'], errors='coerce')

# Encontrar duplicados
duplicados = freeapps[freeapps['App'].duplicated(keep=False)]

# Filtrar para obtener la fila con el máximo número de 'Reviews' por aplicación
max_reviews = duplicados.loc[duplicados.groupby('App')['Reviews'].idxmax()]

# Mostrar el resultado
print("Aplicaciones duplicadas con el número más alto de reseñas:\n", max_reviews)
```

```python
# Verificar que la columna 'Rating' sea numérica
print("Tipo de la columna 'Rating':", freeapps['Rating'].dtype)

# Obtener el valor mínimo y máximo de la columna 'Rating'
min_rating = freeapps['Rating'].min()
max_rating = freeapps['Rating'].max()

# Mostrar los resultados
print("Valor mínimo de 'Rating':", min_rating)
print("Valor máximo de 'Rating':", max_rating)
```

```python

# Función para verificar si una cadena contiene solo caracteres ASCII
def es_ascii(s):
    return all(ord(c) < 128 for c in s)

# Aplicar la función y crear una nueva columna
freeapps['Es_ASCII'] = freeapps['App'].apply(es_ascii)

# Contar cuántas apps tienen solo caracteres ASCII y cuántas no
total_ascii = freeapps['Es_ASCII'].sum()
total_no_ascii = freeapps.shape[0] - total_ascii

# Mostrar resultados
print("Número de aplicaciones con caracteres ASCII:", total_ascii)
print("Número de aplicaciones sin caracteres ASCII:", total_no_ascii)
```

```python
freeappsENG = freeapps[freeapps['Es_ASCII']]
```

```python
freeappsENG.head(10)
```

```python
freeappsENG.shape
```

```python
# 1. Crear tabla de frecuencia para 'Genres'
tabla_genres = freeappsENG['Genres'].value_counts(normalize=True) * 100  # Porcentajes
tabla_genres = tabla_genres.reset_index()
tabla_genres.columns = ['Género', 'Porcentaje']
tabla_genres = tabla_genres.sort_values(by='Porcentaje', ascending=False)  # Ordenar de manera descendente

# 2. Crear tabla de frecuencia para 'Category'
tabla_categoria = freeappsENG['Category'].value_counts(normalize=True) * 100  # Porcentajes
tabla_categoria = tabla_categoria.reset_index()
tabla_categoria.columns = ['Categoría', 'Porcentaje']
tabla_categoria = tabla_categoria.sort_values(by='Porcentaje', ascending=False)  # Ordenar de manera descendente

# Mostrar las tablas de frecuencia
print("Tabla de Frecuencia de Géneros:")
print(tabla_genres)
print("\nTabla de Frecuencia de Categorías:")
print(tabla_categoria)
```

```python
# Crear tabla de frecuencia para 'Installs' (porcentajes)
tabla_installs = freeappsENG['Installs'].value_counts(normalize=True) * 100
tabla_installs = tabla_installs.reset_index()
tabla_installs.columns = ['Installs', 'Porcentaje']

# Ordenar de forma descendente por el porcentaje
tabla_installs = tabla_installs.sort_values(by='Porcentaje', ascending=False)

# Mostrar la tabla de frecuencia
print("Tabla de Frecuencia de 'Installs':")
print(tabla_installs)
```

```python
# Paso 1: Limpiar la columna 'Installs' (remover comas y '+') y convertir a float
freeappsENG['Installs'] = freeappsENG['Installs'].str.replace('[+,]', '', regex=True).astype(float)

# Paso 2: Función para calcular el promedio de instalaciones por género (Genres o Category)
def promedio_installs_por_categoria(df, categoria_col):
    # Agrupar por la columna de categoría y calcular el promedio de 'Installs'
    promedio_installs = df.groupby(categoria_col)['Installs'].mean().reset_index()
    promedio_installs.columns = [categoria_col, 'Promedio Installs']
    
    # Ordenar de mayor a menor el promedio de instalaciones
    promedio_installs = promedio_installs.sort_values(by='Promedio Installs', ascending=False)
    
    return promedio_installs

# Paso 3: Calcular el promedio por 'Genres'
promedio_installs_genres = promedio_installs_por_categoria(freeappsENG, 'Genres')

# Paso 4: Calcular el promedio por 'Category'
promedio_installs_categoria = promedio_installs_por_categoria(freeappsENG, 'Category')

# Mostrar los resultados
print("Promedio de Installs por Género (Genres):")
print(promedio_installs_genres)

print("\nPromedio de Installs por Categoría (Category):")
print(promedio_installs_categoria)
```

```python
# Ajustar el formato para que los números se vean con comas y sin notación científica
pd.options.display.float_format = '{:,.0f}'.format

# Volver a mostrar las tablas con el nuevo formato
print("Promedio de Installs por Género (Genres):")
print(promedio_installs_genres)

print("\nPromedio de Installs por Categoría (Category):")
print(promedio_installs_categoria)
```
